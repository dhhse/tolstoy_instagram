```python
import pandas as pd
```

```python
tolstoj = pd.read_csv('/Users/aleksandranisimov/Desktop/tolstoj/tolstoj.tsv', encoding="utf-8", sep="\t")
```

```python
new_tolstoj = list(tolstoj["Название, датировка, материал, техника,размеры"])
#list = []
```

```python
list_tolstoj = []
for element in new_tolstoj:
    list_tolstoj.append(element.split("\n"))
```

```python
nazvanie = []
for element in list_tolstoj:
    nazvanie.append(element[0])
```

```python
author = {}
for i, element in enumerate(list_tolstoj):
    for el in element:
        if el.startswith("Автор:"):
            author.update({el:[i]})
```

```python
razmery = []
for element in list_tolstoj:
    razmery.append(element[-1])
```

```python
tolstoj_df = pd.DataFrame()
```

```python
authors_df = pd.DataFrame.from_dict(author)
```

```python
authors_df = pd.melt(authors_df)
```

```python
for i, element in enumerate(list_tolstoj):
    tolstoj_df = tolstoj_df.append(pd.DataFrame(
    [{"название":nazvanie[i],
     "размер":razmery[i], 
      "value":i}]),
    sort = True)
```

```python
new_tolstoj = pd.merge(tolstoj_df, authors_df, how = "outer", on = 'value', left_index = False)
```

```python
new_tolstoj = new_tolstoj.drop(columns="value")
```

```python
new_tolstoj = new_tolstoj.fillna(0)
```

```python
new_tolstoj.to_csv(path_or_buf="/Users/aleksandranisimov/Desktop/tolstoj/new_tolstoj.tsv", sep="\t")
```
